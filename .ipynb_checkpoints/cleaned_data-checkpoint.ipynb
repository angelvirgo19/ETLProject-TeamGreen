```python
# Dependencies
import pandas as pd
```

```python
# Read the file
file= "Resources/pollution_us_2000_2016.csv"
df = pd.read_csv(file, encoding="ISO-8859-1")
df.head(2)
```

```python
# Check the columns names
#list(df.columns)
```

```python
# Check null rows
#df.count()
```

```python
# Clean unwanted columns
df2=df.drop(['Unnamed: 0','State Code','County Code','Site Num', 'Address', 'County', 'City',
        'O3 Units', 'O3 Mean', 'O3 1st Max Value','O3 1st Max Hour', 'O3 AQI',
        'NO2 1st Max Value','NO2 1st Max Hour','NO2 AQI',
        'SO2 1st Max Value', 'SO2 1st Max Hour', 'SO2 AQI', 
        'CO 1st Max Value', 'CO 1st Max Hour', 'CO AQI'], axis = 1) 

df2.head(5)
```

```python
# Split the year column and delete part of (%m-%d)
split_year_df = df2["Date Local"].str.split("-", n = 1, expand = True)
df2["Year"]= split_year_df[0]
df2["month-day"]= split_year_df[1]
del df2["month-day"]
df2.head()
```

```python
#Check the rows
#unique = df2["Date Local"].unique()
#print(unique)
#Note: All rows are unique
```

```python
# Create a new dataframe to aggregate the Mean columns.
grouped_df = df2.groupby(["Year"])
mean_df=pd.DataFrame(grouped_df["NO2 Mean", "SO2 Mean", "CO Mean"].mean().reset_index())
#print(mean_df)
mean_df.head(10)
#Note: I tried this methot, but it didn't work.
```

```python
# Order the columns and make a new df
ordered_df=df2[["Year", "State", "NO2 Units", "NO2 Mean", "SO2 Units", "SO2 Mean", "CO Units", "CO Mean"]]
ordered_df.head()
```
