```python
#import dependencies

import pandas as pd
```

```python
#set path and read csv into a DataFrame

path = "Resources/air_emissons_by_source.csv"
df=pd.read_csv(path)
df.head()
```

```python
#print column names

for col in df.columns:
    print(col)
```

```python
#parse down df to only keep the columns we need 
df=df[["COU", "POL", "YEA", "Unit", "Value", "Variable", "VAR"]]
df.head()
```

```python
#remove rows with unnecessary data
pollutants_extra=df[(df['POL']=="NMVOC") | (df['POL']=="PM10") | (df['POL']=="PM2-5")].index
df=df.drop(pollutants_extra)
df.head()
```

```python
#rename values so they will match the other df 
# df['POL'].replace({'SOX': 'SO2', 'NOX': 'NO2'})
df['POL']= df['POL'].replace({'SOX': 'SO2', 'NOX': 'NO2'})
df.head(25)
```

```python
#drop any rows with null data
clean_df = df.dropna(axis = 0, how ='any')
clean_df.head(25)
```

```python
#parse down df
clean_df=clean_df[["POL", "Value", "Variable", "VAR", "YEA"]]
clean_df.head()
```

```python
#rename columns
clean_df=clean_df.rename(columns={"YEA": "Year", "VAR": "pol_var", "Variable": "pol_variable"})
clean_df.head()
```

```python
#so2 df
so2_df=clean_df[(clean_df["POL"]=="SO2")]
so2_df.head(17)
```

```python
#co df
co_df=clean_df[(clean_df["POL"]=="CO")]
co_df.head(17)
```

```python
#no2 df
no2_df=clean_df[(clean_df["POL"]=="NO2")]
no2_df.head(17)
```

```python
#write to individual csv files
so2_df.to_csv("Resources/cleaned_air_emission_by_source_so2.csv", index=False)
co_df.to_csv("Resources/cleaned_air_emission_by_source_co.csv", index=False)
no2_df.to_csv("Resources/cleaned_air_emission_by_source_no2.csv", index=False)
```
