```python
#import dependencies

import pandas as pd
```

```python
#set path and read csv into a DataFrame

path = "Resources/air_emissons_by_source.csv"
df=pd.read_csv(path)
df.head()
```

```python
#print column names

for col in df.columns:
    print(col)
```

```python
#parse down df to only keep the columns we need 
df=df[["COU", "POL", "YEA", "Unit", "Value", "VAR"]]
df.head()
```

```python
#remove rows with unnecessary data
pollutants_extra=df[(df['POL']=="NMVOC") | (df['POL']=="PM10") | (df['POL']=="PM2-5")].index
df=df.drop(pollutants_extra)
df.head()
```

```python
#only keep data with TOT Varaiable 

df=df.loc[df["VAR"]=="TOT"]
df.head(25)
```

```python
#rename values so they will match the other df 
# df['POL'].replace({'SOX': 'SO2', 'NOX': 'NO2'})
df['POL']= df['POL'].replace({'SOX': 'SO2', 'NOX': 'NO2'})
df.head()
```

```python
#set index to year
df=df.set_index(["YEA"])
df.head()
```

```python
clean_df = df.dropna(axis = 0, how ='any')
clean_df.head(25)
```

```python
so2_df=clean_df[(clean_df["POL"]=="SO2")]
so2_df.head(17)
```

```python
co_df=clean_df[(clean_df["POL"]=="CO")]
co_df.head(17)
```

```python
no2_df=clean_df[(clean_df["POL"]=="NO2")]
no2_df.head(17)
```

```python
so2_df.to_csv("cleaned_air_emission_by_source_so2.csv")
co_df.to_csv("cleaned_air_emission_by_source_so2.csv")
no2_df.to_csv("cleaned_air_emission_by_source_so2.csv")
```
